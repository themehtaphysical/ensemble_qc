```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R(omega, phi, Omega, T2):
    R = ((1/T2)*np.cos(phi) + (omega - Omega)*np.sin(phi))/((omega - Omega)**2 + 1/(T2*T2))

    return R

def I(omega, phi, Omega, T2):
    R = ((1/T2)*np.sin(phi) - (omega - Omega)*np.cos(phi))/((omega - Omega)**2 + 1/(T2*T2))

    return R
```

```python
omega_vals = np.arange(-5,5, 0.1)
T2 = 1
Omega = 0
phi = np.arange(0, np.pi + np.pi/4, np.pi/4)
phi_2 = np.arange(-np.pi,0, np.pi/4)

fig, subplot = plt.subplots(4, figsize = (30,17))


for _phi,__phi in zip(phi,phi_2):
    subplot[0].plot(omega_vals, R(omega_vals, _phi, Omega, T2), label="$\\phi =  $"+str(_phi/np.pi) + "$\\times \pi$")
    subplot[1].plot(omega_vals, R(omega_vals, __phi, Omega, T2), label="$\\phi =  $"+str(__phi/np.pi) + "$\\times \pi$")
subplot[0].legend()
subplot[0].set_xlabel("$\\omega$")
subplot[0].set_ylabel("Real(S($\\omega$))")
subplot[0].set_title("real spectrum of FID for different posiitve phi")
subplot[0].axhline(y=0, color='k')
subplot[0].axvline(x=0, color='k')

subplot[1].legend()
subplot[1].set_xlabel("$\\omega$")
subplot[1].set_ylabel("Real(S($\\omega$))")
subplot[1].set_title("real spectrum of FID for different negative phi ")
subplot[1].axhline(y=0, color='k')
subplot[1].axvline(x=0, color='k')

for _phi,__phi in zip(phi,phi_2):
    subplot[2].plot(omega_vals, R(omega_vals, _phi, Omega, T2), label="$\\phi =  $"+str(_phi/np.pi) + "$\\times \pi$")
    subplot[3].plot(omega_vals, R(omega_vals, __phi, Omega, T2), label="$\\phi =  $"+str(__phi/np.pi) + "$\\times \pi$")
subplot[2].legend()
subplot[2].set_xlabel("$\\omega$")
subplot[2].set_ylabel("Img(S($\\omega$))")
subplot[2].set_title("imag spectrum of FID for different posiitve phi")
subplot[2].axhline(y=0, color='k')
subplot[2].axvline(x=0, color='k')

subplot[3].legend()
subplot[3].set_xlabel("$\\omega$")
subplot[3].set_ylabel("Img(S($\\omega$))")
subplot[3].set_title("imag spectrum of FID for different negative phi ")
subplot[3].axhline(y=0, color='k')
subplot[3].axvline(x=0, color='k')


fig.tight_layout()
```
